<h1><span style="font-family:Arial,Helvetica,sans-serif">Optimization Model</span></h1>

<ul>
	<li>
	<p><span style="font-family:Arial,Helvetica,sans-serif"><span style="font-size:20px">The model was implemented on a simplified integrated infrastructure network consisting of water, power, and transportation components.</span></span></p>
	</li>
	<li>
	<p><span style="font-family:Arial,Helvetica,sans-serif"><span style="font-size:20px">Four components in the network were failed (a water pump (P_MP1), an electric&nbsp;line (P_L2), an electric&nbsp;load (P_LO1),&nbsp;and an electric motor connected to the water pump (W_P22).</span></span></p>
	</li>
	<img src="figures/integrated_network.png" alt="Alt text" width="650" title="Title text" />
	<br>
  <hr />
	<li>
	<p><span style="font-family:Arial,Helvetica,sans-serif"><span style="font-size:20px">The objective of the model is to identify the optimal repair order and schedule so as to maximize the resilience of the integrated network.</span></span></p>
	</li>
</ul>
